```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

2. Задати вершини довільного опуклого 8-кутника на площині. Зменшити його в 1,5 рази.

```python
octagon = np.array([[6, 3], [5, 5], [3, 6],
                   [1, 5], [0, 3], [1, 1], 
                   [3, 0], [5, 1], [6, 3]])

# Зменшення 8-кутника в 1.5 рази (тобто масштабування на 2/3)
scaling_factor = 2 / 3
scaling_matrix = np.array([
    [scaling_factor, 0],
    [0, scaling_factor]
])

# Масштабування вершин
scaled_points = np.inner(octagon, scaling_matrix)
```

```python
plt.plot(octagon[:, 0], octagon[:, 1], ".-", label="Original")
plt.plot(scaled_points[:, 0], scaled_points[:, 1], ".-", label="Scaled")
plt.grid(True)
plt.xlim(-8, 8), plt.ylim(-8, 8)
plt.legend()
plt.show()
```

3. Отриманий результат з попереднього пункту симетрично відобразити
відносно початку координат.

```python
mirrored_scaled_points = scaled_points * -1

# Візуалізація результату та його симетричного відображення
plt.plot(scaled_points[:, 0], scaled_points[:, 1], ".-", label="Scaled")
plt.plot(mirrored_scaled_points[:, 0], mirrored_scaled_points[:, 1], ".-", label="Mirrored Scaled")
plt.grid(True)
plt.xlim(-8, 8), plt.ylim(-8, 8)
plt.legend()
plt.show()
```

4. Задати похилу пряму двома точками на площині. Симетрично
відобразити 8-кутник відносно даної прямої. Вказати перетворення, які
для цього необхідні.

```python
point1 = np.array([2, -2]) # Задаємо точки похилої
point2 = np.array([1, -1])

def reflect_point_over_line(point, A, B, C):
    x, y = point
    d = (A * x + B * y + C) / (A**2 + B**2) # Формула відстані від точки до похилої
    x_reflected = x - 2 * A * d
    y_reflected = y - 2 * B * d
    return np.array([x_reflected, y_reflected])

A = point2[1] - point1[1]
B = point1[0] - point2[0]
C = point2[0] * point1[1] - point1[0] * point2[1]

reflected_octagon = np.array([reflect_point_over_line(point, A, B, C) for point in octagon])
```

```python
plt.axline(point1, point2, color='k', linestyle='--', label='Line of Symmetry')
plt.plot(octagon[:, 0], octagon[:, 1], ".-", label="Original Octagon")
plt.plot(reflected_octagon[:, 0], reflected_octagon[:, 1], ".-", label="Mirrored Octagon")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Mirroring Octagon with respect to a Given Line')
plt.grid(True)
plt.xlim(-10, 10), plt.ylim(-10, 10)
plt.legend()
plt.show()
```

6. Виконати пункти 2-4 для одного будь-якого графічного об'єкту, який подано в xlsx-фійлі.

```python
file_path = "var3.xlsx"
data = pd.read_excel(file_path)
#print(data)
```

2. Відобразити графічний об'єкт. Зменшити його в 1,5 рази.

```python
scaling_factor = 2 / 3
scaled_data = data.copy()
scaled_data['x'] = data['x'] * scaling_factor
scaled_data['y'] = data['y'] * scaling_factor
```

```python
plt.figure(figsize=(8, 6))
plt.plot(data['x'], data['y'], "o-", label="Original Data from Excel")
plt.plot(scaled_data['x'], scaled_data['y'], "s--", label="Scaled Data from Excel")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Original and Scaled Data')
plt.grid(True)
plt.legend()
plt.axis('equal')  # Збереження однакових пропорцій по обох осях
plt.show()
```

3. Отриманий результат з попереднього пункту симетрично відобразити
відносно початку координат.

```python
mirrored_scaled_data = scaled_data * -1

plt.plot(scaled_data['x'], scaled_data['y'], "o-", label="Scaled Data from Excel")
plt.plot(mirrored_scaled_data['x'], mirrored_scaled_data['y'], "s--", label="Mirrored Data from Excel")
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
